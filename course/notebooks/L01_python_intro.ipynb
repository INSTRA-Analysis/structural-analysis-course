## 1.1 Variables — Giving Numbers a Name

In Python, a **variable** is a named storage box. You put a value in with `=`.

````python
# Rectangular cross-section dimensions
b = 200    # width  in mm
d = 450    # depth  in mm

print(b)
print(d)
````

Notice the `#` character — anything after it is a **comment** for the human reader. Python ignores it completely.

Now let's calculate the cross-sectional area:

````python
A = b * d          # mm²
print(A)
````

Python's arithmetic operators are exactly what you'd expect:

| Operator | Meaning         | Example         |
|----------|----------------|-----------------|
| `+`      | Addition        | `3 + 2` → `5`  |
| `-`      | Subtraction     | `10 - 4` → `6` |
| `*`      | Multiplication  | `6 * 7` → `42` |
| `/`      | Division        | `9 / 4` → `2.25` |
| `**`     | Power / Exponent| `3 ** 2` → `9` |

Let's calculate the second moment of area (moment of inertia):

````python
I = (b * d**3) / 12    # mm⁴
print(I)
````

That big number is hard to read. Let's display it properly.

---

## 1.2 f-strings — Pretty Results

An **f-string** lets you embed variable values inside a sentence. Put an `f` before the quotes, then use `{}` around any variable name:

````python
print(f"Width  b = {b} mm")
print(f"Depth  d = {d} mm")
print(f"Area   A = {A} mm²")
````

You can also control how many decimal places to show. Add `:.2f` inside the braces — the `2` means two decimal places, `f` means a floating-point number:

````python
Z = I / (d / 2)    # section modulus in mm³

print(f"Second moment of area  I = {I:,.0f} mm⁴")
print(f"Section modulus        Z = {Z:,.0f} mm³")
````

The `,` in `{I:,.0f}` adds thousand-separator commas. Much more readable!

---

## 1.3 Making Decisions with `if / else`

Code can make decisions based on conditions. The structure is:

```python
if <condition is True>:
    # do this
else:
    # do that instead
```

The **indentation** (spaces at the start of the line) is mandatory — it tells Python which lines belong inside the `if`.

````python
# Check that the section is deeper than it is wide (good practice for beams)
if d > b:
    print(f"OK: The section is deeper than it is wide (d={d} > b={b}).")
else:
    print(f"WARNING: The section is wider than it is deep (b={b} > d={d}). Check your input!")
````

Comparison operators for use inside `if` conditions:

| Operator | Meaning                  |
|----------|--------------------------|
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |

## 1.4 Your First Interactive Widget

Now the fun begins. Instead of editing the numbers in the code and re-running, let's create **sliders** that update the calculation live.

We use a library called `ipywidgets`. A **library** is a collection of ready-made tools — you bring it into your code with `import`:

````python
import ipywidgets as widgets
from IPython.display import display
````

Now here is the full interactive widget — **read through each comment before running it**:

````python
# --- Step 1: Create an output area where results will be displayed ---
out = widgets.Output()

# --- Step 2: Create sliders ---
slider_b = widgets.IntSlider(
    value=200, min=100, max=600, step=10,
    description="Width b (mm):",
    style={"description_width": "140px"},
    layout=widgets.Layout(width="420px")
)

slider_d = widgets.IntSlider(
    value=450, min=100, max=900, step=10,
    description="Depth d (mm):",
    style={"description_width": "140px"},
    layout=widgets.Layout(width="420px")
)

# --- Step 3: Define the update function ---
# This function runs every time a slider moves.
def update(change=None):
    b = slider_b.value
    d = slider_d.value

    A = b * d
    I = (b * d**3) / 12
    Z = I / (d / 2)
    slenderness = d / b

    check = "✓ OK" if d >= b else "⚠ Consider increasing depth"

    with out:
        out.clear_output(wait=True)   # clear previous result
        print("─" * 42)
        print(f"  Cross-section: {b} mm × {d} mm")
        print("─" * 42)
        print(f"  Area               A = {A:>12,.0f} mm²")
        print(f"  Second moment      I = {I:>12,.0f} mm⁴")
        print(f"  Section modulus    Z = {Z:>12,.0f} mm³")
        print(f"  Slenderness      d/b = {slenderness:>12.2f}")
        print(f"  Depth check          {check}")
        print("─" * 42)

# --- Step 4: Connect sliders to the update function ---
slider_b.observe(update, names="value")
slider_d.observe(update, names="value")

# --- Step 5: Run once to show initial results, then display everything ---
update()
display(widgets.VBox([slider_b, slider_d, out]))
````

Try moving the sliders. Watch the numbers update instantly without re-running any cell.

This exact pattern — **create output area → create controls → define update function → connect → display** — will appear in every lesson. By Lesson 4 you'll write it from memory.
